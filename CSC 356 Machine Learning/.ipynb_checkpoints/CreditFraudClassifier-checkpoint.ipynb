```python
import numpy as np
import pandas as pd
import sys
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import sklearn
# Machine learning libraries
from sklearn.metrics import classification_report, accuracy_score
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.mixture import GaussianMixture
from sklearn.covariance import EllipticEnvelope
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("creditcard.csv") # Grab the data
```

```python
print(data.columns)
data.shape
```

```python
# random_state apparently ensures the output will be the same when splitting the data
# Helps produce reproducible results
# Frac precentage returned

# Define random_state

default = 1

#data = data.sample(frac = 0.2, random_state = default)
X_train, X_test = train_test_split(data, test_size = 0.2, random_state = 42)
print(X_train.shape)
print(X_test.shape)
print(X_train.columns)
print(X_test.columns)
```

```python
X_train.hist(figsize = (20,20)) # plot a histogram of each parameter
plt.show()
```

```python
# Data clusters around either 0 or 1 usually, according to the source of the data, fraud == 1 and valid == 0
fraudTrain = X_train[X_train["Class"] == 1]
validTrain = X_train[X_train["Class"] == 0]

fraudTest = X_test[X_test["Class"] == 1]
validTest = X_test[X_test["Class"] == 0]

# Calculate amount of outliers
outlier_fraction_train = len(fraudTrain) / float(len(validTrain))
print(outlier_fraction_train)

outlier_fraction_test = len(fraudTest) / float(len(validTest))
print(outlier_fraction_test)

print(f"Valid training cases: {len(validTrain)}")
print(f"Fraud training cases: {len(fraudTrain)}")
print(f"Valid testing cases: {len(validTest)}")
print(f"Fraud testing cases: {len(fraudTest)}")
```

```python
# Create a correlation with X_train
correlation_matrix = X_train.corr()
fig = plt.figure(figsize = (12, 9))

# Use sns to display the data
sns.heatmap(correlation_matrix, vmax = 0.8, square = True)
plt.show() # Lighter squares signify correlation
```

```python
# Organize the data
columns = data.columns.tolist()

# Filter columns to remove unwanted data
columns = [c for c in columns if c not in ['Class']]

target = "Class"

# Put all desired data in X matrix
# X = data[columns]
xTest = X_test[columns]
xTrain = X_train[columns]
# Y is all the labels
# Y = data[target]
yTest = X_test[target]
yTrain = X_train[target]

# Print the shapes
print(xTrain.shape)
print(yTrain.shape)
print(xTest.shape)
print(yTest.shape)
```

```python
classifiers = {
    # contamination is the suspected number of outliers
    "Isolation Forest" : IsolationForest(max_samples = len(xTrain),
                                       contamination = outlier_fraction_train,
                                       random_state = default),
    # Number of neighbors to consider, the higher the percentage of outliers the higher you want to make this number (?)
    "Local Outlier Factor" : LocalOutlierFactor(
        n_neighbors = 20,
        contamination = outlier_fraction_train, novelty = True),
    "Gaussian Mixture" : GaussianMixture(n_components=2, n_init=10),
    "FastMCD" : EllipticEnvelope(support_fraction = 0.9, contamination = outlier_fraction_train, random_state = 42)
}
```

```python
n_outliers_train = len(fraudTrain)

n_outliers_test = len(fraudTest)

# Train the classifiers and use them to predict on the training set!

for i, (clf_name, clf) in enumerate(classifiers.items()): # enumerate gives 2 loop variables instead of 1
    if clf_name == "Local Outlier Factor":
        clf.fit(xTrain)
        y_predict_test = clf.predict(xTest)
        scores_pred_test = clf.negative_outlier_factor_
    elif clf_name == "Gaussian Mixture" or clf_name == "FastMCD":
        clf.fit(xTrain)
        y_predict_train = clf.predict(xTrain)
        scores_pred_train = clf.score_samples(xTrain)
    else:
        clf.fit(xTrain)
        scores_pred_train = clf.decision_function(xTrain)
        y_predict_train = clf.predict(xTrain)
    
    # Turn the prediction value into 0 or 1
    y_predict_train[y_predict_train == 1] = 0
    y_predict_train[y_predict_train == -1] = 1
    
    # Calculate error number
    n_err_train = (y_predict_train != yTrain).sum()
    
    print(f"{clf_name} {n_err_train}")
    print(f"accuracy with train set: {accuracy_score(yTrain, y_predict_train)}")
    print(f"classification report : \n{classification_report(yTrain, y_predict_train)}")
```

```python
# Now run it with the test set!
for i, (clf_name, clf) in enumerate(classifiers.items()): # enumerate gives 2 loop variables instead of 1
    if clf_name == "Local Outlier Factor":
        clf.fit(xTrain)
        y_predict_test = clf.predict(xTest)
        scores_pred_test = clf.negative_outlier_factor_
    elif clf_name == "Gaussian Mixture" or clf_name == "FastMCD":
        clf.fit(xTrain)
        y_predict_test = clf.predict(xTest)
        scores_pred_test = clf.score_samples(xTest)
    else:
        clf.fit(xTrain)
        scores_pred_test = clf.decision_function(xTest)
        y_predict_test = clf.predict(xTest)
    
    # Turn the prediction value into 0 or 1
    y_predict_test[y_predict_test == 1] = 0
    y_predict_test[y_predict_test == -1] = 1
    
    # Calculate error number
    n_err_test = (y_predict_test != yTest).sum()
    
    print(f"{clf_name} {n_err_test}")
    print(f"accuracy with test set: {accuracy_score(yTest, y_predict_test)}")
    print(f"classification report : \n{classification_report(yTest, y_predict_test)}")
```
